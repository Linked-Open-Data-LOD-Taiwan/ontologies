```python
# riversum s2.0
riversrc0_df = read_ods('dataset/riversum.ods',0,0) #中央管區域排水
riversrc11_df = read_ods('dataset/riversum.ods',1,0).drop([None], axis=1).dropna() #中央管河川
riversrc12_df = read_ods('dataset/riversum.ods',2,0).drop([None], axis=1).dropna() #縣市管河川
riversrc13_df = read_ods('dataset/riversum.ods',3,0).drop([None], axis=1).dropna() #跨省市河川

riversrc1_df = pd.concat([riversrc11_df, riversrc12_df,riversrc13_df])
riversrc1_df.reset_index().drop(['index'], axis=1)
#river_df.to_csv('output/riversum_s1.csv')
```

```python
# riversum s2
riversum_df = pd.read_csv('output/riversum_s1.csv')
riversum_colname=['河川代碼','河川名稱','河川英文','流入代號','管理組織代號']
riversum_df.columns=riversum_colname

#排水
riversum_df = pd.merge(riversum_df, riversrc0_df, left_on='河川名稱',right_on='排水路名稱', how='outer')

#xx 管河川
riversum_df = pd.merge(riversum_df, riversrc1_df, left_on='河川名稱',right_on='水系別', how='outer')


# merge into 描述
did=36691
m1df=odMgr.get_dataset(did)
# 河川名詞庫
m2df = m1df[m1df['Category'].str.contains('河川簡介')]
df = pd.merge(riversum_df, m2df, left_on='河川名稱',right_on='ChineseTerminology', how='left')
df=df.drop(['Category', 'ChineseTerminology','EnglishTerminology','SerialNumber'], axis=1)
newcol_name=['描述']
riversum_colname = list(riversum_df.columns)
riversum_colname.extend(newcol_name)
df.columns = riversum_colname

riversum_df = df

riversum_df.to_csv('output/riversum_s2.csv')
riversum_df
#riversum_df.to_clipboard()
```
